```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import adjusted_rand_score

# Parameters
a, b, c, d = 3, 0, 1, 2
s = 0.5
y_true = np.array([0] * 10 + [1] * 10)

# Generate data
x0 = np.hstack([np.random.normal(a, s, 10), np.random.normal(b, s, 10)])
x1 = np.hstack([np.random.normal(c, s, 10), np.random.normal(d, s, 10)])
x = np.vstack([x0, x1]).T

# Plot data with true labels
plt.scatter(x[:, 0], x[:, 1], c=y_true, cmap='viridis')
plt.xlabel('x0')
plt.ylabel('x1')
plt.title("Generated Data Points with True Labels")
plt.show()

# Clustering with K-Means
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans_labels = kmeans.fit_predict(x)
kmeans_centers = kmeans.cluster_centers_

# Clustering with Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=2)
agglo_labels = agglo.fit_predict(x)

# Clustering with DBSCAN
dbscan = DBSCAN(eps=2.0, min_samples=5)
dbscan_labels = dbscan.fit_predict(x)

# Prediction for a new data point [1, 2] using K-Means
new_point_prediction = kmeans.predict([[1, 2]])

# Evaluate clustering quality
kmeans_score = adjusted_rand_score(y_true, kmeans_labels)
agglo_score = adjusted_rand_score(y_true, agglo_labels)
dbscan_score = adjusted_rand_score(y_true, dbscan_labels)

kmeans_labels, kmeans_centers, new_point_prediction, kmeans_score, agglo_labels, agglo_score, dbscan_labels, dbscan_score
```
